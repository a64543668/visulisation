# PLOTTING

```python
# PLOT ONE
def plot_cust(df, title, start_time, end_time):
    
    # Create figure
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    

    trace1 = go.Bar(x=df['start_of_time'], y=df['stop_customer'], marker={'color':'yellow'}, name='a', opacity=0.5)
    trace2 = go.Scatter(x=df['start_of_time'], y=df['total_acut'], marker={'color':'Green'}, name='b')
    trace3 = go.Scatter(x=df['start_of_time'], y=df['total_ivr'], marker={'color':'#00549A'}, name='c')
    
    fig.add_trace(trace1, secondary_y=True)
    fig.add_trace(trace2, secondary_y=False)
    fig.add_trace(trace3, secondary_y=False)
    
    fig.add_shape(type = 'rect',
        xref='x',
        yref='paper',
        x0= start_time,
        y0= 0,
        x1= end_time,
        y1= 1,
        fillcolor= 'red',
        opacity= 0.1,
        line= {
            'color':'red',
            'width': 0,
        })
#     fig.update_layout(title_text=title)
    # Plot the figure
    iplot(fig)
```

```python
for outage in list(df.Outage_Identifier.unique()):
    df_one=df[df.Outage_Identifier == outage]
    df_one.head()
    start_time = df_one['Outage Notify time'].iloc[0]
    end_time = df_one['Outage End time'].iloc[0]
    plot_cust(df_one, outage, start_time, end_time)
```
